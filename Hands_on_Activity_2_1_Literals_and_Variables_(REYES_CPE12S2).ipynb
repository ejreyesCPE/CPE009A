**Procedure:**


Task:

1. A teacher wants to calculate the final grade in a CpE course and want to write it in a python program. The following are the requirements:

  a. PRELIM GRADE = 50% Prelim Exam + 50% Prelim Class Standing (CS)
  
  b. PRELIM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
  
  c. MIDTERM GRADE = 1/3 of PRELIM GRADE + 2/3 of (50% Midterm Exam + 50% Midterm Class Standing (CS))
  
  d. MIDTERM CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
  
  e. FINAL GRADE = 1/3 of MIDTERM GRADE + 2/3 of (50% Final Exam + 50% Final Class Standing (CS))
  
  f. FINAL CS = 50% Hands-on activity + 30% Quiz + 20% Assignment
  
  g. HOAs, Quizzes and Assignments are inputted as average of all submissions and are out of 100%.
  
  h. Major exams are inputted out of 100%.

  i. Show the codes that successfully run the program.

  j. Provide comments or documentation strings for your program.


```python
print("PRELIMS:")
PHOA = float(input("Enter Prelim Hands-on Activity grade: "))
PQUIZ = float(input("Enter Prelim Quiz grade: "))
PAS = float(input("Enter Prelim Assignment grade: "))

# Prelim Class Standing
PCS = (PHOA * 0.5) + (PQUIZ * 0.3) + (PAS * 0.2)
print(f"Prelim CS is: {PCS:.2f}")

PEX = float(input("Enter Prelim Exam grade: "))
# Prelim Grade
PG = (PEX * 0.5) + (PCS * 0.5)
print(f"Prelim Grade is: {PG:.2f}\n")


print("MIDTERMS:")
MHOA = float(input("Enter Midterm Hands-on Activity grade: "))
MQUIZ = float(input("Enter Midterm Quiz grade: "))
MAS = float(input("Enter Midterm Assignment grade: "))

# Midterm Class Standing
MCS = (MHOA * 0.5) + (MQUIZ * 0.3) + (MAS * 0.2)
MEX = float(input("Enter Midterm Exam grade: "))

# Midterm Grade
MG = (1/3 * PG) + (2/3 * ((MEX * 0.5) + (MCS * 0.5)))
print(f"Midterm Grade is: {MG:.2f}\n")


print("FINALS:")
FHOA = float(input("Enter Final Hands-on Activity grade: "))
FQUIZ = float(input("Enter Final Quiz grade: "))
FAS = float(input("Enter Final Assignment grade: "))

# Final Class Standing
FCS = (FHOA * 0.5) + (FQUIZ * 0.3) + (FAS * 0.2)
FEX = float(input("Enter Final Exam grade: "))

# Final Grade
FG = (1/3 * MG) + (2/3 * ((FEX * 0.5) + (FCS * 0.5)))

# Output
print("\n")
print(f"Final Course Grading: {FG:.2f}")
```
