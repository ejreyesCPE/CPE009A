```python
# Assign grade for each subject based on marks:
# 90-100: A+
# 80-89: A
# 70-79: B+
# 60-69: B
# 50-59: C
# 40-49: D
#Below 40: F (Fail)

#Percentage ≥ 90: "Outstanding"
#Percentage ≥ 80: "Excellent"
#Percentage ≥ 70: "Very Good"
#Percentage ≥ 60: "Good"
#Percentage ≥ 40: "Satisfactory"
#Below 40: "Needs Improvement"

def performance (grade):
  if grade >= 90 and grade <= 100:
    print("Mark: A+ | Remarks: Outstanding")
  elif grade >= 80 and grade <= 89:
    print("Mark: A | Remarks: Excellent")
  elif grade >= 70 and grade <= 79:
    print("Mark: B+ | Remarks: Very Good")
  elif grade >= 60 and grade <= 69:
    print("Mark: B | Remarks: Good")
  elif grade >= 50 and grade <= 59:
    print("Mark: C | Remarks: Satisfactory")
  elif grade >= 40 and grade <= 49:
    print("Mark: D | Remarks: Needs Improvement")
  elif grade >= 0 and grade <= 39:
    print("Mark: F | Remarks: Fail")

def overall (total):
  if total >= 90:
    print("Overall Performance: Outstanding")
  elif total >= 80:
    print("Overall Performance: Excellent")
  elif total >= 70:
    print("Overall Performance: Very Good")
  elif total >= 60:
    print("Overall Performance: Good")
  elif total >= 40:
    print("Overall Performance: Satisfactory")
  elif total >= 0:
    print("Overall Performance: Needs Improvement")


subjectcount = int(input("Enter number of subjects: "))
grades_list = []
for i in range(subjectcount):
    subject = print("\nSubject No.", i + 1)
    grade = float(input("Grade: "))
    grades_list.append(grade)
    performance(grade)

total = sum(grades_list) / subjectcount
print("\nTotal Marks: ", total)
overall(total)
```
