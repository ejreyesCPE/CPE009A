```python
# We want to create a python code for a grading scheme (DONE)

# Create a function to accept a number of activities (DONE)
# Compute for the score vs total (DONE)
# Return the average grade of the student in % (DONE)
def termCrit(actNumber):
    rawScore = []
    totalScore = []

    for i in range(0, actNumber):
        print(f"\nActivity {i+1}: ")
        inputScore = float(input("Score: "))
        inputTotal = float(input("Total: "))

        if inputScore > inputTotal:
            print("\nInvalid Scores.\n")
            inputScore, inputTotal = 0, 0
            break
        else:
            rawScore.append(inputScore)
            totalScore.append(inputTotal)

    # Used TIP's new grading scheme as reference
    overallRaw = sum(rawScore)
    overallTotal = sum(totalScore)
    grade = (overallRaw / overallTotal) * 50 + 50
    return grade
```

```python
# Create a function for the class standing per term (DONE)
def termCrit(count):
    total_raw = 0
    total_possible = 0
    for i in range(count):
        raw = int(input(f"Enter score for item {i+1}: "))
        possible = int(input(f"Enter highest possible score for item {i+1}: "))
        total_raw += raw
        total_possible += possible
    5
    if total_possible == 0:
        return 0

    # Based on TIP grading scheme
    percentage = ((total_raw / total_possible) * 50) + 50
    return percentage

# Function for class standing per term
def termCS():
    print("\nHands-on Activities: ")
    termHOA = termCrit(int(input("How many hands-on activities this term? ")))

    print("\nQuizzes:")
    termQuiz = termCrit(int(input("How many quizzes this term? ")))

    print("\nAssignments: ")
    termAss = termCrit(int(input("How many assignments this term? ")))

    grade = (0.5 * termHOA) + (0.3 * termQuiz) + (0.2 * termAss)
    return grade
```

```python
# Create a function to compute the term grade
def termGrade():
  print("Computing for the term grade.")
  examRaw = float(input("Exam Score: "))
  examTotal = float(input("Exam Total: "))
  examGrade = (examRaw/examTotal) * 100

  CSgrade = termCS()

  grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return grade
```

```python
# Create a function to compute for the grades in the semester
def semesterGrade():
  print("Prelim Grade")
  prelimGrade = termGrade()
  print(f"Prelim Grade: {prelimGrade:.2f}")

  print("\nMidterm Grade")
  midtermGrade = termGrade()
  finalMG = (1/3)*prelimGrade + (2/3)*midtermGrade
  print(f"Computed Midterm Grade: {midtermGrade:.2f}")

  print("\nFinal Grade")
  finalGrade = termGrade()
  finalFG = (1/3)*finalMG + (2/3)*finalGrade

  return finalFG
```
