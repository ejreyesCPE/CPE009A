**CONTROL** **STRUCTURES**

If Statement

```python
a = 12
b = 100

if b > a:
  print("b is greater than a")
```

Elif Statements

```python
a = 12
b = 12

if b > a:
  print("b is greater than a")

elif a == b:
  print("a and b are equal")
```

Else Statements

```python
a = 200
b = 33

if b > a:
  print("b is greater than a")

elif a == b:
  print("a and b are equal")

else:
  print("a is greater than b")
```

Short Hand If Statement

```python
if a > b: print("a is greater than b")
```

```python
a = 2
b = 330
print("A") if a > b else print("B")
```

And-both conditions are true

```python
a = 200
b = 33
c = 500
if a > b and c > a:
  print("Both conditions are True")
```

Or-at least one of the conditions is true

```python
a = 200
b = 33
c = 500
if a > b or a > c:
  print("At least one of the conditions is True")
```

Nested If Statements

```python
# Create a python program that will display above 10, above 20, above 30, above 40, but below 50.
# +5 points!
x = 41

if x > 10:
  print("Above ten,")
  if x > 20:
    print("above 20,")
    if x > 30:
      print("above 30,")
      if x > 40:
        print("above 40,")
        if x < 50:
          print("but below 50.")
      else:
        print("not above 40.")
    else:
      print("not above 30.")
  else:
    print("below 20.")
else:
  print("below 10.")
```
